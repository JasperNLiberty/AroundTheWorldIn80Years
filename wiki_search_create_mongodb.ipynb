```python
import pandas as pd
from pymongo import MongoClient
import requests
from bs4 import BeautifulSoup
import json
```

```python
resp = requests.get("http://www.hillmanwonders.com/top_1000/")
soup = BeautifulSoup(resp.content, "lxml")
trs = soup.find("div", id="EdColumn").find("table").find_all("tr")

data = []
for tr in trs:
    tds = tr.find_all("td")
    country_td, names_td = tds[0], tds[1]
    data.extend(names_td.get_text().strip().split('\n'))

query_list = data[:-14]
#with open('saved_dest.html', 'w') as f:
#    f.write(resp.content)
```

```python
# connect to mongodb
client = MongoClient()
db = client.get_database("attractions")
```

```python
# initialize collection
db.get_collection('result')
```

```python
problem_ix = []
for ix, query in enumerate(query_list):
    print('saving ' + str(ix) + '...')
    try:
        description = wiki_search(query)
        result = {'location': query, 'details':description}
        db.result.insert_one(result)
    except:
        problem_ix.append(ix)
```
